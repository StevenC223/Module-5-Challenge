### Analysis

- It looks that Capomulin and Ramicane are effective at reducing tumor size, increases metastatic sites, but it is less effective with a higher weight
- While Ketapril, Naftisol, Stelasyn, Zoniferol and Propriva has 0 or negative effect on the tumor growth site, seems to be what could be described as a placebo effect, and Zoniferol encourages tumor increase.

Overall, a heavier weight had a lesser effect of treatment.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
from scipy.stats import linregress

sns.set()
sns.set_context("notebook")
sns.set_style("whitegrid")

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
mice_number = merge_df["Mouse ID"].unique().size
mice_number
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = merge_df[merge_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mice=merge_df.loc[merge_df["Mouse ID"] == "g989", :]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df[merge_df["Mouse ID"] != "g989"]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_cleaned = clean_df["Mouse ID"].unique().size
mice_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
regimen_group = clean_df.groupby(clean_df["Drug Regimen"])

mice_mean = regimen_group["Tumor Volume (mm3)"].mean()
mice_median = regimen_group["Tumor Volume (mm3)"].median()
mice_variance = regimen_group["Tumor Volume (mm3)"].var()
mice_std = regimen_group["Tumor Volume (mm3)"].std()
mice_sem = regimen_group["Tumor Volume (mm3)"].sem()
# Assemble the resulting series into a single summary DataFrame.
stats_summary = pd.DataFrame ({ "Mean Tumor Volume" : mice_mean,
                                "Median Tumor Volume" : mice_median,
                                "Tumor Volume Variance": mice_variance,
                                "Tumor Volume Standard Deviation" : mice_std,
                                "Tumor Volume Standard Error" : mice_sem
                              }).style.set_caption("Summary Stats for Tumour Volume (mm3)")
stats_summary                        
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line


stats_aggregated = clean_df[["Tumor Volume (mm3)", "Drug Regimen"]].groupby(["Drug Regimen"]).agg(["mean","median","var","std","sem"]).style.set_caption("Summary Statistics")
stats_aggregated
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

max_timepoints = clean_df.groupby(["Mouse ID"])["Timepoint"].max()

max_merge = pd.merge(max_timepoints, clean_df, on=("Mouse ID", "Timepoint"))
max_merge.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
treatments = ["Capomulin","Ramicane", "Infubinol", "Ceftamin"]
capomulin_data = []
ramicane_data = []
infubinol_data = []
ceftamin_data = []

for index, row in max_merge.iterrows():
    if row["Drug Regimen"] == treatments[0]:
        capomulin_data.append(row["Tumor Volume (mm3)"])
    if row["Drug Regimen"] == treatments[1]:
        ramicane_data.append(row['Tumor Volume (mm3)'])
    if row["Drug Regimen"] == treatments[2]:
        infubinol_data.append(row['Tumor Volume (mm3)'])
    if row["Drug Regimen"] == treatments[3]:
        ceftamin_data.append(row['Tumor Volume (mm3)'])


treatments_df = pd.DataFrame ({treatments[0] : capomulin_data,
                               treatments[1] : ramicane_data,
                               treatments[2] : infubinol_data,
                               treatments[3] : ceftamin_data
                              })

treatments_df.head()

# Calculate the IQR and quantitatively determine if there are any potential outliers.
#For Capomulin
cap_q1 = np.percentile(capomulin_data, 25, interpolation = 'midpoint')
cap_q2 = np.percentile(capomulin_data, 50, interpolation = 'midpoint') 
cap_q3 = np.percentile(capomulin_data, 75, interpolation = 'midpoint') 
cap_iqr = cap_q3 - cap_q1
cap_low = cap_q1 - (1.5 * cap_iqr)
cap_up = cap_q3 + (1.5 * cap_iqr)
#For Ramicane
ram_q1 = np.percentile(ramicane_data, 25, interpolation = 'midpoint')
ram_q2 = np.percentile(ramicane_data, 50, interpolation = 'midpoint') 
ram_q3 = np.percentile(ramicane_data, 75, interpolation = 'midpoint') 
ram_iqr = ram_q3 - ram_q1
ram_low = ram_q1 - (1.5 * ram_iqr)
ram_up = ram_q3 + (1.5 * ram_iqr)
#For Infubinol
inf_q1 = np.percentile(infubinol_data, 25, interpolation = 'midpoint')
inf_q2 = np.percentile(infubinol_data, 50, interpolation = 'midpoint') 
inf_q3 = np.percentile(infubinol_data, 75, interpolation = 'midpoint') 
inf_iqr = inf_q3 - inf_q1
inf_low = inf_q1 - (1.5 * inf_iqr)
inf_up = inf_q3 + (1.5 * inf_iqr)
#For Ceftamin
cef_q1 = np.percentile(ceftamin_data, 25, interpolation = 'midpoint')
cef_q2 = np.percentile(ceftamin_data, 50, interpolation = 'midpoint') 
cef_q3 = np.percentile(ceftamin_data, 75, interpolation = 'midpoint') 
cef_iqr = cef_q3 - cef_q1
cef_low = cef_q1 - (1.5 * cef_iqr)
cef_up = cef_q3 + (1.5 * cef_iqr)

    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset
    # Determine outliers using upper and lower bounds
outlier_cap = []
outlier_ram = []
outlier_inf = []
outlier_cef = []

for x in capomulin_data:
    if((x>cap_up) or (x<cap_low)):
        outlier_cap.append(x)
for x in ramicane_data:
    if((x>ram_up) or (x<ram_low)):
        outlier_ram.append(x)
for x in infubinol_data:
    if((x>inf_up) or (x<inf_low)):
        outlier_inf.append(x)
for x in ceftamin_data:
    if((x>cef_up) or (x<cef_low)):
        outlier_cef.append(x)
#Output
print("Outlier in the Capomulin dataset are", outlier_cap)
print("Outlier in the Ramicane dataset are", outlier_ram)
print("Outlier in the Infubinol dataset are", outlier_inf)
print("Outlier in the Ceftamin dataset are", outlier_cef)
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
outliers = dict(marker='x', markerfacecolor = 'green', markersize=12)
plot_list = [treatments_df[treatments[0]],
             treatments_df[treatments[1]],
             treatments_df[treatments[2]],
             treatments_df[treatments[3]]]

fig1, ax= plt.subplots(figsize=(9,7))
ax.set_title("Final Tumor Sizes by Drug")
ax.set_xlabel("Drug Regimen")
ax.set_ylabel("Tumor Size (mm3)")
ax.boxplot(plot_list, patch_artist=True, boxprops=dict(facecolor="red"), showmeans=False, flierprops=outliers)
plt.xticks([1,2,3,4],treatments)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
campoulin_df = clean_df[["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]] 
campoulin_df = campoulin_df.groupby("Mouse ID").mean()
average_weight = campoulin_df["Weight (g)"]
average_volume = campoulin_df["Tumor Volume (mm3)"]

plt.scatter(campoulin_df["Weight (g)"], campoulin_df["Tumor Volume (mm3)"], marker="x", facecolors="red")
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Tumor vs Mouse Weight - Capomulin")


plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
co_co = st.pearsonr(average_weight, average_volume)
print(f"The correlation between mouse weight and average tumour volume is {co_co}")

reg_model = linregress(average_weight,average_volume)
(slope, intercept, rvalue, pvalue, stderr) = reg_model
co_plot = average_weight * slope + intercept



line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(average_weight, average_volume, marker = "x", facecolors="blue")
plt.plot(average_weight, co_plot, color="red")


plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Size (mm3)")
plt.title("Average Tumor vs Mouse weight w/ Capomulin", fontsize=20)
plt.annotate(line_eq, (24.5,34),fontsize=14,color="blue")
plt.show
```
